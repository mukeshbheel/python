```python
## independent and dependent features
X = df[['Weight']] #independent freature should be dataframe or two dimentional array
y = df['Height'] # this variable can be in series or one dimentional array

# type(df) # type is dataframe
# type(X) #type is series, series means just a single record if X = df['Weight']
```

```python
X_series = df['Weight']
np.array(X_series).shape #type is series, series means just a single record if X = df['Weight']
```

```python
## Train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
x_train.shape
```

```python
## standardization
# we only standardrize independent features because we only use independent features in gradient desent

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# we use fit transform for training data only, and for testing data we use transform, then the transform uses the same u and 
# standard devialtion value for test data also, if we use fit_transform for test data also then it will use the u and standard deviation
# for test data transformation causing data leakage
## we dont want test data to know any info about train data so to avoid data leakage we use transform only for test data
```

```python
## Applying Simple Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train, y_train)
# regression.fit(X_series, y) gives error because X_series is an 1-D array
```
